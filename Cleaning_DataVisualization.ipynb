## Removing the duplicate rows

```python
df_raw = df.drop_duplicates()
df_raw.shape
```

## Handling missing values

```python
for i in df_raw.columns:
    if df_raw[i].isna().sum() != 0:
        print(f"{i} : {df_raw[i].isna().sum()}")
```

```python
# Dropping the columns which are having more than 40000 nan values
df_raw.drop(['SHOOTING','PACE','ATT_POSITION','LOAN_DATE_END'],axis = 1,inplace = True)
```

```python
# Replacing the categorical values with mode and numeric values with mean 

for i in df_raw.columns:
    if df_raw[i].dtype == object:
        df_raw[i] = df_raw[i].fillna(df_raw[i].mode()[0])
    else:
        df_raw[i] = df_raw[i].fillna(df_raw[i].mean())
```

## Correcting the data types

```python
df_raw.dtypes
```

```python
# now we will print the unique values for each column non numeric column
for i in df_raw.columns:
    if df_raw[i].dtype == object:
        print(i)
        print(df_raw[i].unique())
        print(f"total: {len(df_raw[i].unique())}")
        print("="*100)
```

```python
df_raw.drop(['NAME','TEAM_CONTRACT'],axis = 1,inplace = True)
```

```python
# Sample column
# df_raw['OVA'] = ...

# Step 1: Clean the values
df_raw['OVA'] = (
    df_raw['OVA']
    .astype(str)                         # Convert all to string
    .str.replace(r'\n', '', regex=True)  # Remove newline characters
    .str.replace(r'n.*', '', regex=True) # Remove everything after 'n' (like n+1, n-1)
    .str.replace(r'\D', '', regex=True)  # Remove any remaining non-digit characters
)

# Step 2: Convert to integer
df_raw['OVA'] = pd.to_numeric(df_raw['OVA'], errors='coerce')

# Optional: Check result
print(df_raw['OVA'].unique())
```

```python
# Sample column
# df_raw['OVA'] = ...

# Step 1: Clean the values
df_raw['POT'] = (
    df_raw['POT']
    .astype(str)                         # Convert all to string
    .str.replace(r'\n', '', regex=True)  # Remove newline characters
    .str.replace(r'n.*', '', regex=True) # Remove everything after 'n' (like n+1, n-1)
    .str.replace(r'\D', '', regex=True)  # Remove any remaining non-digit characters
)

# Step 2: Convert to integer
df_raw['POT'] = pd.to_numeric(df_raw['POT'], errors='coerce')

# Optional: Check result
print(df_raw['POT'].unique())
```

```python
for i in df_raw['HEIGHT']:
    i = i.split('cm')[0]
```

```python
# Step 1: Extract the 'cm' portion using regex
df_raw['HEIGHT'] = (
    df_raw['HEIGHT']
    .str.extract(r'(\d+)cm')   # Extract digits before 'cm'
    .astype(float)             # Convert to numeric type
    .astype('Int64')           # Optional: convert to integer with support for NaNs
)

# Step 2: (Optional) Check the result
print(df_raw['HEIGHT'].unique())
```

```python
# Step 1: Extract the number before 'lbs'
df_raw['WEIGHT'] = (
    df_raw['WEIGHT']
    .str.extract(r'(\d+)lbs')   # Extract digits before 'lbs'
    .astype(float)              # Convert to numeric type
    .astype('Int64')            # Optional: keep integers with NaN support
)

# Step 2: (Optional) Check the result
print(df_raw['WEIGHT'].unique())
```

# converting these columns into integer types JOINED VALUE WAGE RC

```python
# no need modify the JOINED column so we are dropping it
df_raw.drop('JOINED',axis = 1,inplace = True)
```

```python
def convert_value(val):
    val = val.replace('€', '').strip()
    
    if val.endswith('M'):
        return int(float(val.replace('M', '')) * 1_000_000)
    elif val.endswith('K'):
        return int(float(val.replace('K', '')) * 1_000)
    elif val == '0':
        return 0
    else:
        return pd.NA  # For unexpected formats

# Apply the function to create a new clean column
df_raw['VALUE'] = df_raw['VALUE'].apply(convert_value)

# Optional: Check unique cleaned values
print(df_raw['VALUE'].unique())
```

```python
def convert_value(wage):
    wage = wage.replace('€', '').strip()
    
    if wage.endswith('M'):
        return int(float(wage.replace('M', '')) * 1_000_000)
    elif wage.endswith('K'):
        return int(float(wage.replace('K', '')) * 1_000)
    elif wage == '0':
        return 0
    else:
        return pd.NA  # For unexpected formats

# Apply the function to create a new clean column
df_raw['WAGE'] = df_raw['WAGE'].apply(convert_value)
df_raw['WAGE'] = pd.to_numeric(df_raw['WAGE'], errors='coerce'); df_raw.dropna(subset=['WAGE'], inplace=True)


# Optional: Check unique cleaned values
print(df_raw['WAGE'].unique())
```
