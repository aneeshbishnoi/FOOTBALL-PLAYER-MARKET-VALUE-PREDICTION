## Converting non numerical data into numerical

```python

for col in ['AGE', 'OVA', 'POT', 'GROWTH', 'BOV']:
    df_copy[col] = pd.to_numeric(df_copy[col], errors='coerce')
    df_copy[col] = df_copy[col].fillna(df_copy[col].median()).astype(int)
```

## Encoding data 

```python
from sklearn.preprocessing import LabelEncoder

bp_encoder = LabelEncoder()
df_copy['BP'] = bp_encoder.fit_transform(df_copy['BP'].astype(str))
```

```python
df_copy['FOOT'] = df_copy['FOOT'].map({'Right': 0, 'Left': 1})
df_copy['FOOT'] = df_copy['FOOT'].fillna(0) 
```

```python
def money_to_numeric(money_str):
    if isinstance(money_str, str):
        money_str = money_str.replace('€', '').strip()
        if 'M' in money_str:
            return float(money_str.replace('M', '')) * 1_000_000
        elif 'K' in money_str:
            return float(money_str.replace('K', '')) * 1_000
        else:
            return float(money_str)
    return pd.NA

df_copy['RC'] = df_copy['RC'].apply(money_to_numeric)
df_copy['RC'] = df_copy['RC'].fillna(df_copy['RC'].median())
```

```python
skill_total_cols = [col for col in df_copy.columns if col.startswith('SKILL_') or col.startswith('TOTAL_')]
for col in skill_total_cols:
    df_copy[col] = pd.to_numeric(df_copy[col], errors='coerce')
    df_copy[col] = df_copy[col].fillna(df_copy[col].median())
```

```python
tech_cols = ['CROSSING', 'FINISHING', 'HEADING_ACCURACY', 'SHORT_PASSING', 'VOLLEYS']
for col in tech_cols:
    df_copy[col] = pd.to_numeric(df_copy[col], errors='coerce')
    df_copy[col] = df_copy[col].fillna(df_copy[col].median())
```

```python
phys_cols = ['ACCELERATION', 'SPRINT_SPEED', 'AGILITY', 'REACTIONS']
for col in phys_cols:
    df_copy[col] = pd.to_numeric(df_copy[col], errors='coerce')
    df_copy[col] = df_copy[col].fillna(df_copy[col].median())
```

```python
def_cols = ['DEFENDING', 'PHYSICAL', 'STRENGTH', 'STAMINA']
for col in def_cols:
    df_copy[col] = pd.to_numeric(df_copy[col], errors='coerce')
    df_copy[col] = df_copy[col].fillna(df_copy[col].median())
```

```python
gk_cols = [col for col in df_copy.columns if col.startswith('GK_')]
for col in gk_cols:
    df_copy[col] = pd.to_numeric(df_copy[col], errors='coerce')
    df_copy[col] = df_copy[col].fillna(df_copy[col].median())
```

```python
# Split contract years into start and end
df_copy[['CONTRACT_START', 'CONTRACT_END']] = df_copy['TEAM_CONTRACT'].str.extract(r'(\d{4})\s*-\s*(\d{4})')
df_copy['CONTRACT_START'] = pd.to_numeric(df_copy['CONTRACT_START'], errors='coerce')
df_copy['CONTRACT_END'] = pd.to_numeric(df_copy['CONTRACT_END'], errors='coerce')
```

### Fill missing international reputation with median

```python

df_copy['INTERNATIONAL_REPUTATION'] = pd.to_numeric(df_copy['INTERNATIONAL_REPUTATION'], errors='coerce')
df_copy['INTERNATIONAL_REPUTATION'] = df_copy['INTERNATIONAL_REPUTATION'].fillna(
    df_copy['INTERNATIONAL_REPUTATION'].median()
)

# Label encode BODY_TYPE
from sklearn.preprocessing import LabelEncoder
body_encoder = LabelEncoder()
df_copy['BODY_TYPE'] = body_encoder.fit_transform(df_copy['BODY_TYPE'].astype(str))
```

```python
def money_to_numeric(money_str):
    if isinstance(money_str, str):
        money_str = money_str.replace('€', '').strip()
        if 'M' in money_str:
            return float(money_str.replace('M', '')) * 1_000_000
        elif 'K' in money_str:
            return float(money_str.replace('K', '')) * 1_000
        else:
            return float(money_str)
    return np.nan

df_copy['VALUE'] = df_copy['VALUE'].apply(money_to_numeric)
df_copy['WAGE'] = df_copy['WAGE'].apply(money_to_numeric)

numeric_cols = df_copy.select_dtypes(include=[np.number]).columns
df_copy[numeric_cols] = df_copy[numeric_cols].fillna(df_copy[numeric_cols].median())


df_copy['JOINED'] = pd.to_datetime(df_copy['JOINED'], errors='coerce')

df_copy.drop(columns=['INTERNATIONAL_REPUTATION', 'CONTRACT_START', 'CONTRACT_END'], inplace=True)

df_copy['ATTACK_WORK_RATE'] = df_copy['ATTACK_WORK_RATE'].astype(str)
df_copy['DEFENSIVE_WORK_RATE'] = df_copy['DEFENSIVE_WORK_RATE'].astype(str)
```

## Feature Engineering 

```python
df_copy['GROWTH_POTENTIAL'] = df_copy['POT'] - df_copy['OVA']
```

```python
df_copy['PHYSICAL_POWER'] = (
    df_copy['STAMINA'] + df_copy['STRENGTH'] + df_copy['JUMPING']
)
```

```python
df_copy['SKILL_INTENSITY'] = (
    df_copy['CROSSING'] + df_copy['FINISHING'] + df_copy['DRIBBLING'] +
    df_copy['BALL_CONTROL'] + df_copy['SHORT_PASSING']
)
```

```python
df_copy['DEFENSE_SCORE'] = (
    df_copy['DEFENSIVE_AWARENESS'] + df_copy['STANDING_TACKLE'] + df_copy['SLIDING_TACKLE']
)
```

```python
df_copy['ATTACKING_MINDSET'] = (
    df_copy['VISION'] + df_copy['COMPOSURE'] + df_copy['AGGRESSION']
)
```

```python
df_copy['EXPLOSIVENESS'] = (
    df_copy['ACCELERATION'] + df_copy['SPRINT_SPEED'] + df_copy['AGILITY']
)
```

```python
df_copy['VALUE_WAGE_RATIO'] = df_copy['VALUE'] / (df_copy['WAGE'] + 1)
```

### Splitting the data into Training and Testing

```python
x = df_copy.drop('VALUE',axis = 1)
y = df_copy['VALUE']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

### Random Forest

```python
def RF():
    from sklearn.ensemble import RandomForestRegressor

    model = RandomForestRegressor(
        n_estimators=50,         
        max_depth=5,             
        min_samples_leaf=10,     
        max_features='sqrt',     
        random_state=42          
    )

    model.fit(x_train, y_train)
    return model
```

### K Nearest Neighbour Regression

```python
from sklearn.neighbors import KNeighborsRegressor
def KNN():
    model = KNeighborsRegressor()
    model.fit(x_train,y_train)
    return model
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

def DT():
    model = DecisionTreeRegressor(
        criterion='squared_error',
        max_depth=3,             
        min_samples_split=10,    
        min_samples_leaf=5,      
        max_features=0.5,        
        random_state=42
    )
    model.fit(x_train, y_train)
    return model
```

### Ridge Regression

```python
from sklearn.linear_model import Ridge

def create_ridge_model():
    model = Ridge(alpha=1.0)
    model.fit(x_train, y_train)
    return model
```

### Model Evaluation function

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score

def evaluate_model(model):
    y_pred = model.predict(x_test)
    MAE = mean_absolute_error(y_test, y_pred)
    MSE = mean_squared_error(y_test, y_pred)
    RMSE = root_mean_squared_error(y_test, y_pred)
    R2 = r2_score(y_test, y_pred)
    return f"{MAE:.2f}", f"{MSE:.2f}", f"{RMSE:.2f}", f"{R2:.2f}"
```

```python
models = []
models.append({"name": "Random Forest", "model": RF()})
models.append({"name": "Decision Tree", "model": DT()})
models.append({"name": "Ridge Regression", "model": create_ridge_model()})
models.append({"name": "K Nearest Neighbor Regressor","model":KNN()})
```

```python
# evaluate all the models
for info in models:
    MAE, MSE, RMSE, R2 = evaluate_model(info["model"])
    info["Mean Absolute Error (MAE)"] = MAE
    info["Mean Squared Error (MSE)"] = MSE
    info["Root Mean Squared Error (RMSE)"] = RMSE
    info["R2 – Score"] = R2
```

```python
# print the evaluation report
df_evaluation_report = pd.DataFrame(models, columns=["name", "Mean Absolute Error (MAE)", "Mean Squared Error (MSE)", "Root Mean Squared Error (RMSE)", "R2 – Score"])
df_evaluation_report
```
